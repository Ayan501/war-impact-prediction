# load data and library

```python
import pandas as pd
clean_data = pd.read_csv(r"C:\Users\Hp\Desktop\My Work\Prepare before war\Data\Processed_data\war_impact_clean.csv")
```

* check how many datatype is in the dataset 

```python
print(clean_data.info())
print(clean_data.dtypes)
```

* convert columns 'gwno_a', 'gwno_loc', 'region' into numeric feature because they has very low cardinality

```python
clean_data['gwno_loc'] = clean_data['gwno_loc'].str.split(",").str[0]
clean_data['gwno_a'] = clean_data['gwno_a'].str.split(",").str[0]
clean_data['region'] = clean_data['region'].str.split(",").str[0]
cols = ['gwno_a','gwno_loc','region']

for col in cols:
    clean_data[col] = pd.to_numeric(clean_data[col],errors='coerce')
```

* Drop country column because country and loaction gives same info 

```python

clean_data.drop(columns=['country'], inplace=True)
```

* check unique value in location columns and remove unnecessary joins


```python
clean_data['location'] = clean_data['location'].str.split(',').str[0]
print(clean_data['location'])
print(clean_data['location'].value_counts())
```

* encode location feature by using frequency encoding

```python
freq = clean_data['location'].value_counts(normalize= True)
print(freq)
clean_data['location_encode'] = clean_data['location'].map(freq)
print(clean_data['location_encode'])
```

* drop location feature

```python
clean_data.drop(columns = 'location', inplace = True)
```

* final check of NaN in dataset

```python
clean_data.isna().sum()
```

* Save this complete data for Model Training

```python
clean_data.to_csv("../Data/Processed_data/Final_clean_dataset.csv", index=False)
print(f"Saved! Shape: {clean_data.shape}")
print(clean_data.columns.tolist())
```
